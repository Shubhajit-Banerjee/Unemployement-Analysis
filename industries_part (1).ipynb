```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('factories_data.csv')
```

```python
dataset.head()
```

# plotting data of States

```python
#plotting data of each state
dataset.plot(legend=True)
plt.legend(bbox_to_anchor=(0.04, 0.82, 1., .102))
plt.show()
```

# Its seems to be  confusing let plot single state data to understand it

1. plotting data of Assam

```python
plt.figure(figsize=(10,10))
plt.plot(dataset.Year,dataset.Assam)
plt.xticks(rotation=60)
plt.title("Assam")
plt.xlabel("Year",labelpad=10)
plt.ylabel("Number of Factories")
plt.show()
```

# Transposing the data

```python
dataset=dataset.transpose()
```

```python
dataset.head(26)
```

# Missing data

```python
X=dataset.iloc[:,:].values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='constant',fill_value=1)
imputer.fit(X[1:, :])
X[1:, :] = imputer.transform(X[1:, :])
```
